```python
def count_numbers(numbers):
  """
  Counts the number of integers in a list that are greater than zero.

  Args:
    numbers: A list of numbers.

  Returns:
    The number of integers in the list that are greater than zero.
  """

  count = 0
  for number in numbers:
    if isinstance(number, int) and number > 0:
      count += 1

  return count

# Example usage
numbers = [1, 2, -3, 0, 4, -5]
positive_count = count_numbers(numbers)
print(f"The number of positive integers in the list is: {positive_count}")
```

# Solution

```python
class Contact:
    def __init__(self, name, phone_number, email=None, address=None):
        self.name = name
        self.phone_number = phone_number
        self.email = email
        self.address = address

    def display(self):
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")
        if self.email:
            print(f"Email: {self.email}")
        if self.address:
            print(f"Address: {self.address}")
        print("-" * 20)

class Phonebook:
    def __init__(self):
        self.contacts = {}

    def add_contact(self, contact):
        if contact.name in self.contacts:
            print(f"Contact with name '{contact.name}' already exists.")
        else:
            self.contacts[contact.name] = contact
            print(f"Contact '{contact.name}' added successfully.")

    def view_contact(self, name):
        if name in self.contacts:
            self.contacts[name].display()
        else:
            print(f"Contact with name '{name}' not found.")

    def list_all_contacts(self):
        if self.contacts:
            print("--- All Contacts ---")
            for name, contact in sorted(self.contacts.items()): # Sort contacts by name
                contact.display()
        else:
            print("Phonebook is empty.")

    def delete_contact(self, name):
        if name in self.contacts:
            del self.contacts[name]
            print(f"Contact '{name}' deleted successfully.")
        else:
            print(f"Contact with name '{name}' not found.")

    def search_contact(self, query):
        results = []
        query = query.lower()
        for name, contact in self.contacts.items():
            if query in name.lower() or \
               query in contact.phone_number or \
               (contact.email and query in contact.email.lower()) or \
               (contact.address and query in contact.address.lower()):
                results.append(contact)

        if results:
            print("--- Search Results ---")
            for contact in sorted(results, key=lambda c: c.name): # Sort search results by name
                contact.display()
        else:
            print(f"No contacts found matching '{query}'.")

    def edit_contact(self, name, new_phone=None, new_email=None, new_address=None):
        if name in self.contacts:
            contact = self.contacts[name]
            if new_phone:
                contact.phone_number = new_phone
                print(f"Phone number for '{name}' updated to '{new_phone}'.")
            if new_email is not None: # Allow setting email to None
                contact.email = new_email
                print(f"Email for '{name}' updated to '{new_email}'.")
            if new_address is not None: # Allow setting address to None
                contact.address = new_address
                print(f"Address for '{name}' updated to '{new_address}'.")
        else:
            print(f"Contact with name '{name}' not found.")

def main():
    my_phonebook = Phonebook()

    while True:
        print("\n--- Phonebook Menu ---")
        print("1. Add Contact")
        print("2. View Contact")
        print("3. List All Contacts")
        print("4. Delete Contact")
        print("5. Search Contacts")
        print("6. Edit Contact")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter name: ")
            phone = input("Enter phone number: ")
            email = input("Enter email (optional): ")
            address = input("Enter address (optional): ")
            new_contact = Contact(name, phone, email, address)
            my_phonebook.add_contact(new_contact)
        elif choice == '2':
            name = input("Enter name to view: ")
            my_phonebook.view_contact(name)
        elif choice == '3':
            my_phonebook.list_all_contacts()
        elif choice == '4':
            name = input("Enter name to delete: ")
            my_phonebook.delete_contact(name)
        elif choice == '5':
            query = input("Enter search query: ")
            my_phonebook.search_contact(query)
        elif choice == '6':
            name = input("Enter name to edit: ")
            if name in my_phonebook.contacts:
                print("Enter new details (leave blank to keep current):")
                new_phone = input(f"New phone number (current: {my_phonebook.contacts[name].phone_number}): ") or None
                new_email = input(f"New email (current: {my_phonebook.contacts[name].email}): ")
                if not new_email:
                    new_email = None
                new_address = input(f"New address (current: {my_phonebook.contacts[name].address}): ")
                if not new_address:
                    new_address = None
                my_phonebook.edit_contact(name, new_phone, new_email, new_address)
            else:
                print(f"Contact with name '{name}' not found.")
        elif choice == '7':
            print("Exiting phonebook.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

# 02_pop_up_shop.md

# Starter Code

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def display(self):
        print(f"  - {self.name}: Rs.{self.price} (In Stock: {self.stock})")

class PopUpShop:
    def __init__(self, name):
        self.name = name
        self.inventory = {}  # Dictionary to store products (name: Product object)
        self.sales = []      # List to store sale records (list of purchased products)

    def add_product(self, product):
        if product.name in self.inventory:
            print(f"Product '{product.name}' already exists in the inventory.")
        else:
            self.inventory[product.name] = product
            print(f"Product '{product.name}' added to the inventory.")

    def display_inventory(self):
        if self.inventory:
            print(f"\n--- {self.name} Inventory ---")
            for product in self.inventory.values():
                product.display()
        else:
            print(f"\n{self.name} inventory is empty.")

    def process_sale(self, cart):
        """
        Processes a sale given a cart (dictionary of product names and quantities).
        Updates stock and records the sale.
        """
        total_price = 0
        sold_items = {}

        for item_name, quantity in cart.items():
            if item_name in self.inventory:
                product = self.inventory[item_name]
                if product.stock >= quantity:
                    product.stock -= quantity
                    total_price += product.price * quantity
                    sold_items[product] = quantity
                else:
                    print(f"Insufficient stock for '{item_name}'. Available: {product.stock}, Requested: {quantity}")
                    return None  # Indicate failed sale
            else:
                print(f"Product '{item_name}' not found in the inventory.")
                return None  # Indicate failed sale

        if sold_items:
            self.sales.append(sold_items)
            print("\n--- Sale Successful ---")
            for product, quantity in sold_items.items():
                print(f"  - {product.name}: {quantity} x Rs.{product.price} = Rs.{product.price * quantity}")
            print(f"Total: Rs.{total_price}")
            return total_price
        else:
            print("\nNo items were sold.")
            return 0

    def display_sales_history(self):
        if self.sales:
            print("\n--- Sales History ---")
            for sale in self.sales:
                sale_total = 0
                print("  --- Sale ---")
                for product, quantity in sale.items():
                    item_total = product.price * quantity
                    print(f"    - {product.name}: {quantity} x Rs.{product.price} = Rs.{item_total}")
                    sale_total += item_total
                print(f"    Total: Rs.{sale_total}")
        else:
            print("\nNo sales recorded yet.")

# Example Usage
if __name__ == "__main__":
    my_shop = PopUpShop("Weekend Wonders")

    # Add products
    my_shop.add_product(Product("T-Shirt", 500, 50))
    my_shop.add_product(Product("Mug", 250, 100))
    my_shop.add_product(Product("Notebook", 150, 75))

    # Display inventory
    my_shop.display_inventory()

    # Process a sale
    customer_cart1 = {"T-Shirt": 2, "Mug": 3}
    my_shop.process_sale(customer_cart1)
    my_shop.display_inventory()

    customer_cart2 = {"Notebook": 5, "Sticker": 10} # Sticker not in inventory
    my_shop.process_sale(customer_cart2)

    customer_cart3 = {"T-Shirt": 55} # Insufficient stock
    my_shop.process_sale(customer_cart3)

    # Display updated inventory
    my_shop.display_inventory()

    # Display sales history
    my_shop.display_sales_history()
```

# Solution

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def display(self):
        print(f"  - {self.name}: Rs.{self.price} (In Stock: {self.stock})")

class PopUpShop:
    def __init__(self, name):
        self.name = name
        self.inventory = {}
        self.sales = []

    def add_product(self, product):
        if product.name in self.inventory:
            print(f"Product '{product.name}' already exists.")
        else:
            self.inventory[product.name] = product
            print(f"Product '{product.name}' added.")

    def display_inventory(self):
        if self.inventory:
            print(f"\n--- {self.name} Inventory ---")
            for product in sorted(self.inventory.values(), key=lambda p: p.name):
                product.display()
        else:
            print(f"\n{self.name} inventory is empty.")

    def process_sale(self, cart):
        total_price = 0
        sold_items = {}

        for item_name, quantity in cart.items():
            if item_name in self.inventory:
                product = self.inventory[item_name]
                if product.stock >= quantity:
                    product.stock -= quantity
                    total_price += product.price * quantity
                    sold_items[product] = quantity
                else:
                    print(f"Insufficient stock for '{item_name}'. Available: {product.stock}, Requested: {quantity}")
                    return None
            else:
                print(f"Product '{item_name}' not found.")
                return None

        if sold_items:
            self.sales.append(sold_items)
            print("\n--- Sale Successful ---")
            for product, quantity in sold_items.items():
                print(f"  - {product.name}: {quantity} x Rs.{product.price} = Rs.{product.price * quantity}")
            print(f"Total: Rs.{total_price}")
            return total_price
        else:
            print("\nNo items were sold.")
            return 0

    def display_sales_history(self):
        if self.sales:
            print("\n--- Sales History ---")
            for i, sale in enumerate(self.sales, 1):
                sale_total = 0
                print(f"  --- Sale #{i} ---")
                for product, quantity in sale.items():
                    item_total = product.price * quantity
                    print(f"    - {product.name}: {quantity} x Rs.{product.price} = Rs.{item_total}")
                    sale_total += item_total
                print(f"    Total: Rs.{sale_total}")
        else:
            print("\nNo sales recorded yet.")

def main():
    shop = PopUpShop("The Corner Store")

    while True:
        print("\n--- Pop-Up Shop Menu ---")
        print("1. Add Product")
        print("2. View Inventory")
        print("3. Process Sale")
        print("4. View Sales History")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter product name: ")
            try:
                price = float(input("Enter product price: "))
                stock = int(input("Enter initial stock: "))
                new_product = Product(name, price, stock)
                shop.add_product(new_product)
            except ValueError:
                print("Invalid input for price or stock. Please enter numbers.")
        elif choice == '2':
            shop.display_inventory()
        elif choice == '3':
            cart = {}
            print("\nEnter items to purchase (enter 'done' when finished):")
            while True:
                item_name = input("Product name: ").strip()
                if item_name.lower() == 'done':
                    break
                try:
                    quantity = int(input(f"Quantity of '{item_name}': "))
                    if quantity > 0:
                        cart[item_name] = quantity
                    else:
                        print("Quantity must be greater than 0.")
                except ValueError:
                    print("Invalid quantity. Please enter a whole number.")

            if cart:
                shop.process_sale(cart)
            else:
                print("Cart is empty.")
        elif choice == '4':
            shop.display_sales_history()
        elif choice == '5':
            print("Thank you for visiting!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
import random
import string

def generate_password(length=12, use_lowercase=True, use_uppercase=True, use_digits=True, use_symbols=True):
    """Generates a strong password with the specified criteria."""

    characters = ""
    if use_lowercase:
        characters += string.ascii_lowercase
    if use_uppercase:
        characters += string.ascii_uppercase
    if use_digits:
        characters += string.digits
    if use_symbols:
        characters += string.punctuation

    if not characters:
        return "Error: At least one character type must be selected."

    password = ''.join(random.choice(characters) for _ in range(length))
    return password

def check_password_strength(password):
    """Checks the strength of a given password based on common criteria."""

    length = len(password)
    has_lowercase = any(c.islower() for c in password)
    has_uppercase = any(c.isupper() for c in password)
    has_digit = any(c.isdigit() for c in password)
    has_symbol = any(c in string.punctuation for c in password)

    strength = 0
    reasons = []

    if length >= 8:
        strength += 1
    else:
        reasons.append("Password should be at least 8 characters long.")

    if has_lowercase:
        strength += 1
    else:
        reasons.append("Password should contain lowercase letters.")

    if has_uppercase:
        strength += 1
    else:
        reasons.append("Password should contain uppercase letters.")

    if has_digit:
        strength += 1
    else:
        reasons.append("Password should contain digits.")

    if has_symbol:
        strength += 1
    else:
        reasons.append("Password should contain symbols.")

    print(f"Password Length: {length}")
    print(f"Contains Lowercase: {has_lowercase}")
    print(f"Contains Uppercase: {has_uppercase}")
    print(f"Contains Digits: {has_digit}")
    print(f"Contains Symbols: {has_symbol}")
    print("-" * 30)

    if strength >= 4 and length >= 10:
        print("Password Strength: Strong")
    elif strength >= 3 and length >= 8:
        print("Password Strength: Medium")
    else:
        print("Password Strength: Weak")
        if reasons:
            print("Suggestions for improvement:")
            for reason in reasons:
                print(f"- {reason}")

# Example Usage
if __name__ == "__main__":
    print("--- Password Generator ---")
    password_length = int(input("Enter desired password length: "))
    include_lowercase = input("Include lowercase letters? (y/n): ").lower() == 'y'
    include_uppercase = input("Include uppercase letters? (y/n): ").lower() == 'y'
    include_digits = input("Include digits? (y/n): ").lower() == 'y'
    include_symbols = input("Include symbols? (y/n): ").lower() == 'y'

    generated_password = generate_password(
        length=password_length,
        use_lowercase=include_lowercase,
        use_uppercase=include_uppercase,
        use_digits=include_digits,
        use_symbols=include_symbols
    )

    print(f"\nGenerated Password: {generated_password}")

    print("\n--- Password Strength Checker ---")
    password_to_check = input("Enter a password to check its strength: ")
    check_password_strength(password_to_check)
```
